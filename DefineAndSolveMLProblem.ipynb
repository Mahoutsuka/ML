```python
features = ['country', 'year', 'Positive affect','Negative affect',
                   'Log GDP per capita', 'Social support',
                   'Healthy life expectancy at birth', 
                   'Freedom to make life choices', 'Generosity', 
                   'Perceptions of corruption', 'Confidence in national government']
label = ['Life Ladder']
features_and_label = features + label
df = df[features_and_label]

renaming = {'Log GDP per capita': 'LogGDP', 
            'Social support': 'Support', 
            'Healthy life expectancy at birth': 'Life', 
            'Freedom to make life choices': 'Freedom', 
            'Perceptions of corruption': 'Corruption',
            'Positive affect': 'Positive', 
            'Negative affect': 'Negative',
            'Confidence in national government': 'Confidence',
            'Generosity': 'Generous'
            }
df = df.rename(renaming, axis=1)
```

```python
#Removing countries that only appear once within the dataset
country_counts = df['country'].value_counts()
filtered_country = country_counts[country_counts>1].index
df = df[df['country'].isin(filtered_country)]
print(df.columns)
```

```python
#Removing NaN values from the dataset
nan_count = df.isnull().sum()
for column,count in nan_count.items():
    #if a column has Nan values, use the mean of the country for that NaN value
    if count > 0:
        df[column] = df[column].fillna(df.groupby('country')[column].transform('mean'))
        # if column still has NaN values, use the world's mean
        if df[column].isnull().sum()>0:
            world_mean = df[column].mean()
            df[column] = df[column].fillna(world_mean)
print(df.isnull().sum())
```

```python
def winsorize_feature(a):
    return stats.mstats.winsorize(a,limits=[0.01,0.01])

for feature in features:
    if feature not in ['country','year']:
        feat = renaming[feature]
        df[feat] = df.groupby('country')[feat].transform(winsorize_feature)
```

```python
#Label for each example shuold be the next year's happiness score
print(df.head(2))
df['Happiness'] = df.groupby('country')['Life Ladder'].shift(-1)
#Drop current year's happiness to avoid data leakage
df.drop(columns = 'Life Ladder', inplace = True)
#Drops the last recored year for each country
df.dropna(inplace = True)
print(df.head(2))
```

```python
#Creates a column identifying the max year for each country 
#This will come in later when creating training and test sets

df['max_year'] = df.groupby('country')['year'].transform('max')
df['second_max_year'] = df.groupby('country')['year'].transform('max') -1

# One-hot encoding of the countries
df_country = pd.get_dummies(df['country'])
df = df.join(df_country)
df = df.drop('country',axis=1)
print(df.columns)

#Creating test and training sets with previous years being in the training set and 
#the most recent year being in the test set
test_rows = df['year'] == df['max_year']
val_rows = df['year'] == df['second_max_year']
train_rows = (~test_rows) & (~val_rows)

test_set = df[test_rows].drop(columns = ['max_year','second_max_year','year'])
val_set = df[val_rows].drop(columns = ['max_year','second_max_year','year'])
train_set = df[train_rows].drop(columns = ['max_year','second_max_year','year'])

X_test = test_set.drop(columns = 'Happiness')
y_test = test_set['Happiness']

X_val = val_set.drop(columns = 'Happiness')
y_val = val_set['Happiness']

X_train = train_set.drop(columns = 'Happiness')
y_train = train_set['Happiness']
```

```python
X_train.head(10)
X_val.head(10)
```

```python
nn_model = keras.Sequential()

input_layer = keras.layers.InputLayer(input_shape = X_train.shape[1])
nn_model.add(input_layer)

hidden_layer_1 = keras.layers.Dense(units = 128, activation = 'relu',kernel_regularizer=keras.regularizers.l2(0.01))
batchNorm_1 = keras.layers.BatchNormalization()
nn_model.add(hidden_layer_1)
nn_model.add(batchNorm_1)


hidden_layer_2 = keras.layers.Dense(units = 64, activation = 'relu',kernel_regularizer=keras.regularizers.l2(0.01))
batchNorm_2 = keras.layers.BatchNormalization()
nn_model.add(hidden_layer_2)
nn_model.add(batchNorm_2)


hidden_layer_3 = keras.layers.Dense(units = 32, activation = 'relu',kernel_regularizer=keras.regularizers.l2(0.01))
batchNorm_3 = keras.layers.BatchNormalization()
nn_model.add(hidden_layer_3)
nn_model.add(batchNorm_3)


hidden_layer_4 = keras.layers.Dense(units = 16, activation = 'relu',kernel_regularizer=keras.regularizers.l2(0.01))
batchNorm_4 = keras.layers.BatchNormalization()
nn_model.add(hidden_layer_4)
nn_model.add(batchNorm_4)

#no activation as life ladder is a continuous scale from 0 - 10
output_layer = keras.layers.Dense(units=1)
nn_model.add(output_layer)

nn_model.summary()
```

```python
sgd_optimizer = keras.optimizers.SGD(learning_rate = 0.001)
loss_fn = keras.losses.MeanSquaredError()
nn_model.compile(optimizer=sgd_optimizer,loss=loss_fn,metrics=[keras.metrics.MeanAbsoluteError()])
```

```python
class ProgBarLoggerNEpochs(keras.callbacks.Callback):
    
    def __init__(self, num_epochs: int, every_n: int = 50):
        self.num_epochs = num_epochs
        self.every_n = every_n
    
    def on_epoch_end(self, epoch, logs=None):
        if (epoch + 1) % self.every_n == 0:
            s = 'Epoch [{}/ {}]'.format(epoch + 1, self.num_epochs)
            logs_s = ['{}: {:.4f}'.format(k.capitalize(), v)
                      for k, v in logs.items()]
            s_list = [s] + logs_s
            print(', '.join(s_list))
```

```python
t0 = time.time()

num_epochs = 250

history = nn_model.fit(X_train,y_train,epochs=num_epochs,verbose=0,callbacks=[ProgBarLoggerNEpochs(num_epochs,every_n=5)],validation_data=(X_val,y_val)) 


t1 = time.time()

print('Elapsed time: %.2fs' % (t1-t0))
```

```python
history.history.keys()
```

```python
# Plot training and validation loss
plt.plot(range(1, num_epochs+ 1), history.history['loss'], label='Training Loss')
plt.plot(range(1, num_epochs+1), history.history['val_loss'], label='Validation Loss')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()


# Plot training and validation accuracy
plt.plot(range(1, num_epochs+1), history.history['mean_absolute_error'], label='Training Error')
plt.plot(range(1, num_epochs+1), history.history['val_mean_absolute_error'], label='Validation Error')

plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
#Comparing the model's prediction to the actual value
happiness_predictions = nn_model.predict(X_test)
for i in range (20):
    print(happiness_predictions[i], y_test.to_numpy()[i])
```

During training, I observed that the validation error fluctuated more compared to the training error, with sudden spikes reaching as high as 714. Despite these fluctuations, the final training and validation losses were very close, suggesting that there were no obvious signs of overfitting.

The sudden spikes in validation loss could be attributed to countries with minimal data and having only 2 or 3 years of records. For these countries, the model likely struggled to generalize well, resulting in poorer performance and increased error on certain validation samples.

Due to the temporal nature of the dataset, implementing grid search with cross-validation was not possible without risking data leakage, as cross-validation randomly splits the data without preserving time order. Therefore, hyperparameter tuning (including learning rate, epochs, and L2 regularization) was conducted manually by repeatedly re-running the model.

Originally, I wanted to include both batch normalization and dropout layers in the MLP to decrease overfitting and improve generalization. However, adding both caused the model to break and be unable to make predictions, so I only kept batch normalization.
